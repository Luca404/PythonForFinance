```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
#Procter&Gamble, Microsoft, Ford and General Electric
tickers = ['PG', 'MSFT', 'F', 'GE']
data = pd.DataFrame()
for ticker in tickers:
    data[ticker] = wb.DataReader(ticker, data_source="yahoo", start='2005-1-1')['Adj Close']
```

```python
returns = np.log(data/data.shift(1))
returns
```

```python
pesi = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
pFolioVar = np.dot(pesi.T, np.dot(returns.cov() * 250, pesi))
pFolioVar * 100
```

```python
pFolioVolatility = (np.dot(pesi.T, np.dot(returns.cov() * 250, pesi))) ** 0.5
pFolioVolatility * 100
```
