```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ["MSFT", "^IXIC"]
df = pd.DataFrame()
for ticker in tickers:
    df[ticker] = wb.DataReader(ticker, data_source="yahoo", start="2017-1-1", end="2021-12-31")["Adj Close"]
```

```python
returns = np.log( df / df.shift(1) )
```

```python
cov = returns.cov() * 250
cov
```

```python
covWithMarket = cov.iloc[0,1]
covWithMarket
```

```python
marketVariance = returns["^IXIC"].var() * 250
marketVariance
```

```python
beta = covWithMarket / marketVariance
beta
```
