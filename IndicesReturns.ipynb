```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^N225']
indData = pd.DataFrame()
for ticker in tickers:
    indData[ticker] = wb.DataReader(ticker, data_source="yahoo", start='2002-1-4')['Adj Close']
```

```python
indData.tail()
```

```python
indData.head()
```

```python
(indData / indData.iloc[0] * 100).plot(figsize=(17,10)); 
plt.show()
```

```python
indReturns = (indData / indData.shift(1) ) - 1
indReturns.tail()
```

```python
annualReturns = indReturns.mean() * 250
annualReturns
```
