```python
import pandas as pd
import numpy as np

from scipy import stats
import statsmodels.api as sm

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/regressionData.csv")
```

```python
data.tail()
```

```python
corrMatrix = data.corr()
print( round(corrMatrix["CPI"][1],3))
```

```python
X = data["UnempRate"]
Y = data["CPI"]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X, Y)
plt.axis([2, 15, -2, 10])
plt.ylabel("CPI Rate")
plt.xlabel("Unemployment Rate")
plt.show()
```

```python
X1 = sm.add_constant(X)

reg = sm.OLS(Y, X1).fit()
```

```python
reg.summary()
```

```python
res = stats.linregress(X,Y)
```

```python
plt.scatter(X,Y)
plt.ylabel("CPI Rate")
plt.xlabel("Unemployment Rate")
plt.plot(X, res.intercept + res.slope*X, 'r', label='fitted line')
plt.show()
```
