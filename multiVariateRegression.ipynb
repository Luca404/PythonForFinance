```python
import pandas as pd
import numpy as np

from scipy import stats
import statsmodels.api as sm

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/multivariateRegressionData.csv")
del data["Date"]
data.head()
```

```python
X = data[["UnempRate", "Commodities"]]
Y = data["CPI"]
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()
reg.summary()
```
